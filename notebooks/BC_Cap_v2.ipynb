```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
coverage = pd.read_csv('../data/2015-2020_ACS_Insurance.csv',  encoding='cp1252')
```

```python
coverage
```

```python
coverage.columns = coverage.columns.str.strip()
```

```python
coverage.info()
```

```python
cov = coverage[['county','Year','Total – Insured Population (Female)','Total – Uninsured Population (Female)']]
```

```python
cov
```

```python
cov.columns = ['county', 'year', 'insured_female', 'uninsured_female']
cov
```

```python
cov.groupby(['county']).insured_female.agg({'sum'})
```

```python
cov.groupby(['county']).agg({'insured_female': ['sum']})
```

```python
insured = cov.groupby(['county']).agg({'insured_female': ['sum']})
insured
```

```python
cov.groupby(['county']).agg({'uninsured_female': ['sum']})
```

```python
uninsured = cov.groupby(['county']).agg({'uninsured_female': ['sum']})
uninsured
```

```python
insurance = insured.merge(uninsured, on =['county'], how='outer')
insurance
```

```python
death_rate = pd.read_csv('../data/2015-2020 death.csv')
death_rate
```

```python
death_rate = death_rate.rename(columns={'County' : 'county'})
death_rate
```

```python
bc_rate = bc_rate.rename(columns={'Race' : 'race'})
bc_rate
```

```python
bc_rate.drop(bc_rate.index[bc_rate['county']=='Tennessee'], inplace=True)
```

```python
bc_rate
```

```python
bc_rate.drop(bc_rate.index[bc_rate['county']=='United States'], inplace=True)
```

```python
sns.barplot(data=top15, x='county', y='(uninsured_female, sum)')
```

```python
plt.figure(figsize=(6, 6))

corr = top15.corr()

# create a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, cmap=cmap, mask = mask, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
